```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Likelihood function for Bernoulli(Î¸) with 3 heads and 2 tails
theta_vals = np.linspace(0.01, 0.99, 500)
likelihood = theta_vals**3 * (1 - theta_vals)**2

# Plot the likelihood function
plt.plot(theta_vals, likelihood, label=r'$L(\theta) = \theta^3 (1 - \theta)^2$')
plt.axvline(x=0.6, color='red', linestyle='--', label=r'MLE $\hat{\theta} = 0.6$')
plt.title('Likelihood Function for Bernoulli Trials')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$L(\theta)$')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/bernoulli_mle.pdf')
```
