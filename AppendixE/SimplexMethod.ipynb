```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the figure and axis
fig, ax = plt.subplots()

# Define constraint boundaries
x = np.linspace(0, 4.5, 300)
y1 = 4 - x          # x1 + x2 <= 4
y2 = np.full_like(x, 3)  # x2 <= 3
x2_limit = np.full_like(x, 2)  # x1 <= 2

# Fill feasible region
x_fill = [0, 0, 1, 2, 2]
y_fill = [0, 3, 3, 2, 0]
ax.fill(x_fill, y_fill, color='lightblue', alpha=0.5, label='Feasible Region')

# Plot constraints
ax.plot(x, y1, label='$x_1 + x_2 = 4$', color='black')
ax.plot(x, y2, label='$x_2 = 3$', color='black')
ax.axvline(2, label='$x_1 = 2$', color='black')

# Plot simplex path
path = np.array([[0, 0], [0, 3], [1, 3], [2, 2]])
ax.plot(path[:, 0], path[:, 1], color='red', linewidth=2, marker='o', label='Simplex Path')
for i in range(len(path) - 1):
    ax.annotate('', xy=path[i + 1], xytext=path[i],
                arrowprops=dict(edgecolor='blue', arrowstyle='->', lw=2))

# Plot objective direction
ax.arrow(0, 0, 0.5, 1, head_width=0.08, fc='k', ec='k', linestyle='dashed')
ax.text(0.6, 1.1, 'increasing $x_1 + 2x_2$', fontsize=10)

# Annotate optimal point
ax.text(2.1, 1.95, 'Optimal', fontsize=10)

# Plot formatting
ax.set_xlim(-0.05, 4.05)
ax.set_ylim(-0.05, 4)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.grid(alpha=0.5)
ax.legend(loc='upper right')

plt.savefig('figures/simplex_method.pdf')
```
