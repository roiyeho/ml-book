```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up the x-axis range
x = np.linspace(-1.5, 1.5, 400)

# Convex functions with different curvature magnitudes
# Small curvature: f_flat(x) = 0.2 x^2
f_flat = lambda x: 0.3 * x**2
# Large curvature: f_sharp(x) = x^2
f_sharp = lambda x: 2 * x**2

# Choose a point where we draw the tangent
x0 = -0.6

# Tangent for small-curvature function
y0_flat = f_flat(x0)
slope_flat = 0.6 * x0  # derivative of 0.2 x^2 is 0.4 x
tangent_flat = y0_flat + slope_flat * (x - x0)

# Tangent for large-curvature function
y0_sharp = f_sharp(x0)
slope_sharp = 4 * x0   # derivative of x^2 is 2x
tangent_sharp = y0_sharp + slope_sharp * (x - x0)

# Set up subplots: two panels, shared y-axis
fig, axes = plt.subplots(1, 2, figsize=(10, 3.5), sharey=True)

# Plot: Small curvature (flat region)
axes[0].plot(x, f_flat(x), label=r'$f(x)$')
axes[0].plot(x, tangent_flat, linestyle='--', color='r', label='Tangent')
axes[0].set_title("Small curvature (flat region)", fontsize=12)
axes[0].set_xlabel(r"$x$", fontsize=12)
axes[0].set_ylabel(r"$f(x)$", fontsize=12)
axes[0].set_xlim([-1, 1])
axes[0].set_ylim([0, 1.5])
axes[0].grid(alpha=0.5)
axes[0].legend()

# Plot: Large curvature (sharp region)
axes[1].plot(x, f_sharp(x), label=r'$f(x)$')
axes[1].plot(x, tangent_sharp, linestyle='--', color='r', label='Tangent')
axes[1].set_title("Large curvature (sharp region)", fontsize=12)
axes[1].set_xlabel(r"$x$", fontsize=12)
axes[1].set_xlim([-1, 1])
axes[1].set_ylim([0, 1.5])
axes[1].grid(alpha=0.5)
axes[1].legend()

plt.savefig("figures/curvature_effect.pdf", bbox_inches='tight')
```
