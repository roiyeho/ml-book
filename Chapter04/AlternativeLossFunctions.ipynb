```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.5, 1.8, 2.4, 3.5, 4.2, 4.8, 5.8, 6.1, 7.2, 8.7, 10])
y = np.array([0.1, 0.2, 0.3, 0.4, 0.7, 1, 0.9, 1.2, 1.4, 1.8, 10])
```

```python
def plot_data(x, y):
    plt.scatter(x, y)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.grid()
```

```python
plot_data(x, y)
plt.savefig('figures/huber_loss_data.pdf')
```

```python
from sklearn.linear_model import SGDRegressor

X = x.reshape(-1, 1) # Reshape x to a 2D array

reg1 = SGDRegressor(loss='squared_error')
reg1.fit(X, y)

reg2 = SGDRegressor(loss='huber')
reg2.fit(X, y)
```

```python
def plot_regression_line(x, y, w0, w1, color, label):
    p_x = np.array([x.min(), x.max()])
    p_y = w0 + w1 * p_x
    plt.plot(p_x, p_y, color, label=label)
```

```python
plot_data(x, y)
plot_regression_line(x, y, reg1.intercept_, reg1.coef_[0], 'r', 'Squared loss')
plot_regression_line(x, y, reg2.intercept_, reg2.coef_[0], 'm', 'Huber loss')
plt.legend()
plt.savefig('figures/huber_loss_results.pdf')
```
