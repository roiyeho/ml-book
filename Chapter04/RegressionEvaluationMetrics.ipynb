```python
import numpy as np
```

```python
# Generate synthetic data with two features
np.random.seed(42)  # For reproducibility
n = 100  # Number of data points

x1 = 2 * np.random.rand(n)
x2 = 3 * np.random.rand(n)
y = 5 + 1 * x1 + 2 * x2 + np.random.randn(n)
```

```python
from sklearn.linear_model import LinearRegression

X = np.c_[x1, x2]  # Concatenating the two features

model = LinearRegression()
model.fit(X, y)
```

```python
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
from sklearn.metrics import mean_squared_error as MSE

# Generating predictions from the model
y_pred = model.predict(X)

# Setting 'squared=False' computes the square root of MSE directly
rmse = MSE(y, y_pred, squared=False)
print(f'RMSE: {rmse:.4f}')
```

```python
from sklearn.metrics import mean_absolute_error 

mae = mean_absolute_error(y, y_pred)
print(f'MAE: {mae:.4f}')
```

```python
r2_score = model.score(X, y)
print(f'R2 score: {r2_score:.4f}')
```
