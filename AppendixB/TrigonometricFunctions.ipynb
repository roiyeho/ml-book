```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Draw unit circle
circle = plt.Circle((0, 0), 1, color='blue', fill=False, linewidth=2, label=r'Unit Circle: $x^2 + y^2 = 1$')
ax.add_patch(circle)

# Angle theta
theta = np.pi / 4  # 45 degrees

# Compute coordinates on the circle
x = np.cos(theta)
y = np.sin(theta)

# Draw the radius from origin to (cosθ, sinθ)
ax.plot([0, x], [0, y], 'r', linewidth=2, label=r'Radius ($r = 1$)')

# Draw projection lines
ax.plot([x, x], [0, y], 'm--', linewidth=1)
ax.plot([0, x], [y, y], 'g--', linewidth=1)

# Mark point and label coordinates
ax.scatter([x], [y], color='black', zorder=3)
ax.text(x + 0.05, y + 0.05, r'$(\cos \theta, \sin \theta)$', fontsize=12, color='black', ha='left', va='bottom')

# Label projections
ax.text(x, -0.1, r'$\cos \theta$', fontsize=12, color='magenta', ha='center')
ax.text(-0.05, y, r'$\sin \theta$', fontsize=12, color='green', ha='right', va='center')

# Draw angle arc
arc_theta = np.linspace(0, theta, 100)
arc_radius = 0.3
ax.plot(arc_radius * np.cos(arc_theta), arc_radius * np.sin(arc_theta), color='red', linewidth=1)
ax.text(0.32, 0.08, r'$\theta$', fontsize=12, color='red')

# Add x and y axes
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Add axis labels
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)

# Set plot limits and appearance
ax.set_xlim(-1.2, 1.4)
ax.set_ylim(-1.2, 1.4)
ax.grid(True, alpha=0.5)
ax.legend(loc='upper left', fontsize=10)

plt.savefig('figures/trigonometric_functions.pdf')
```
