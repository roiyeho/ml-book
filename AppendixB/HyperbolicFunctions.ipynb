```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Plot the unit hyperbola x^2 - y^2 = 1 for x >= 1 and x <= -1
x = np.linspace(1, 3, 300)
y = np.sqrt(x**2 - 1)
ax.plot(x, y, 'b', label=r'Unit Hyperbola: $x^2 - y^2 = 1$')
ax.plot(x, -y, 'b')

x_neg = np.linspace(-3, -1, 300)
y_neg = np.sqrt(x_neg**2 - 1)
ax.plot(x_neg, y_neg, 'b')
ax.plot(x_neg, -y_neg, 'b')

# Choose a representative angle theta
theta = np.pi / 4
cosh_theta = np.cosh(theta)
sinh_theta = np.sinh(theta)

# Draw the radius from the origin to (cosh(theta), sinh(theta))
ax.plot([0, cosh_theta], [0, sinh_theta], 'r', linewidth=2)

# Draw projection lines
ax.plot([cosh_theta, cosh_theta], [0, sinh_theta], 'm--', linewidth=1)
ax.plot([0, cosh_theta], [sinh_theta, sinh_theta], 'g--', linewidth=1)

# Draw the hyperbolic angle arc
arc_radius = 0.6
arc = Arc((0, 0), 2*arc_radius, 2*arc_radius, angle=0, theta1=0, theta2=30, color='red')
ax.add_patch(arc)
ax.text(0.65, 0.15, r'$\theta$', fontsize=13, color='red')

# Point label
ax.scatter([cosh_theta], [sinh_theta], color='black', zorder=3)
ax.text(cosh_theta + 0.1, sinh_theta, r'$(\cosh \theta, \sinh \theta)$', fontsize=12, color='black', ha='left', va='bottom')
ax.text(cosh_theta, -0.3, r'$\cosh \theta$', fontsize=12, color='m', ha='center')
ax.text(-0.1, sinh_theta, r'$\sinh \theta$', fontsize=12, color='g', va='center', ha='right')

# Draw axes
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Add axis labels
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)

# Grid and ticks
ax.set_xticks(np.arange(-3, 4, 1))
ax.set_yticks(np.arange(-3, 4, 1))
ax.grid(True, alpha=0.5)

# Appearance
ax.set_aspect('equal')
ax.legend(loc='upper left', fontsize=10)

# Save figure
plt.savefig('figures/hyperbolic_functions.pdf')
```
