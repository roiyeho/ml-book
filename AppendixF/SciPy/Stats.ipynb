```python
import numpy as np

np.random.seed(42)
```

```python
from scipy.stats import norm

# PDF and CDF of the standard normal distribution
pdf = norm.pdf(0)
cdf = norm.cdf(1.96)

# Generate 5 random samples from N(0, 1)
samples = norm.rvs(size=5)

# Compute the 95th percentile (inverse CDF)
quantile = norm.ppf(0.95)

print("PDF(0):", pdf)
print("CDF(1.96):", cdf)
print("Samples:", samples)
print("95th percentile:", quantile)
```

```python
from scipy.stats import ttest_ind

# Simulated test scores for two groups
group1 = [85, 90, 88, 75, 95]
group2 = [78, 82, 84, 80, 76]

t_stat, p_value = ttest_ind(group1, group2)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```
