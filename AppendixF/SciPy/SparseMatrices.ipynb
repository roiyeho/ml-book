```python
import numpy as np
```

```python
from scipy.sparse import csr_matrix

# Dense 2D array with many zeros
A_dense = np.array([
    [0, 0, 1],
    [1, 0, 0],
    [0, 0, 2]
])

# Convert to CSR format
A_sparse = csr_matrix(A_dense)

# Multiply by a dense vector
x = np.array([1, 2, 3])
y = A_sparse.dot(x)

print("Result:", y)
```
