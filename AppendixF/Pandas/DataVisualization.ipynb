```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/flights_sample_3m.csv')
```

Line Plot

```python
df[['CRS_ELAPSED_TIME', 'AIR_TIME']].head(500).plot()
plt.savefig('figures/dataframe_line_plot.pdf')
```

Customizing Plots

```python
ax = df[['CRS_ELAPSED_TIME', 'AIR_TIME']].head(500).plot(
    figsize=(8, 4),
    grid=True,
    color=['tab:blue', 'tab:purple']
)
ax.set_title('Scheduled vs. Actual Flight Duration (First 500 Flights)')
ax.set_xlabel('Flight Index')
ax.set_ylabel('Time (minutes)')
ax.legend(['Scheduled (CRS)', 'Actual Air Time'])

plt.savefig('figures/dataframe_line_plot_custom.pdf')
```

Bar Plot

```python
df['AIRLINE'].value_counts().plot.bar()
plt.savefig('figures/dataframe_bar_plot.pdf', bbox_inches='tight')
```
