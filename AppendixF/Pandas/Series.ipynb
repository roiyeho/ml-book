```python
import pandas as pd
```

Creating Series

```python
s1 = pd.Series([10, 20, 30, 40])
s1
```

```python
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2
```

```python
grades = {'Math': 92, 'English': 88, 'History': 79, 'Science': 94}
s3 = pd.Series(grades)
s3
```

Attributes

```python
s3.shape
```

```python
s3.dtype
```

```python
s3.values
```

Indexing and Slicing

```python
s3.iloc[1]
```

```python
s3['English']
```

```python
s3.iloc[1:3]  # Elements at positions 1 and 2
```

```python
s3['English': 'Science']
```

```python
s3[s3 > 90]
```

```python
s3[['History', 'Math']]
```

Vectorized Operations

```python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

s * 2
```

```python
import numpy as np

np.log(s)
```

```python
s.mean()
```
