```python
import numpy as np
```

```python
x = np.array([1, 2])
y = np.array([3, 4])

np.dot(x, y)  # or x @ y
```

```python
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[4,  2], 
              [-2, 1]])
A @ B
```

```python
np.linalg.inv(A)
```

```python
np.linalg.det(A)
```

```python
np.linalg.solve(A, [1, 0])
```

```python
vals, vecs = np.linalg.eig(A)

print("Eigenvalues:", vals)
print("Eigenvectors:\n", vecs)
```
