```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
x + 10
```

```python
A = np.arange(15).reshape(5, 3)
B = np.array([100, 200, 300])  # shape: (3,)
A + B  # shape: (5, 3), B is broadcast along the first axis
```

```python
A = np.ones((2, 3, 4))
B = np.array([[10], [20], [30]])  # shape: (3, 1)
A + B  # shape: (2, 3, 4), B is broadcast along the first and last axes
```

```python
A = np.ones((3, 2))
B = np.array([10, 20, 30])  # shape: (3,)
A + B
```
