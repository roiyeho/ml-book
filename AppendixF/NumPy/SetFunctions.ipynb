```python
import numpy as np
```

```python
x = np.array([1, 2, 2, 3, 4])
y = np.array([3, 4, 4, 5, 6])

print(np.unique(x))
print(np.intersect1d(x, y))
print(np.union1d(x, y))
print(np.isin(x, y))
```

```python
x = np.array([4, 1, 2, 1, 5, 2, 1])

values, counts = np.unique(x, return_counts=True)

print("Values:", values)
print("Counts:", counts)
```
