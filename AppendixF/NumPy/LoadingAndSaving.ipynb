```python
import numpy as np
```

```python
A = np.array([[1, 2], [3, 4]])
np.save('my_array.npy', A)
```

```python
B = np.load('my_array.npy')
B
```

```python
x = np.arange(5)
y = np.linspace(0, 1, 5)

np.savez('data.npz', x=x, y=y)
loaded = np.load('data.npz')
print('x =', loaded['x'])
print('y =', loaded['y'])
```
