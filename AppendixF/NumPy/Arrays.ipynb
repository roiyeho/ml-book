Memory Layout

```python
A = np.arange(6).reshape(2, 3)
print(f'A.shape: {A.shape}, A.strides: {A.strides}')
```

```python
B = A.T
print(f'B.shape: {B.shape}, B.strides: {B.strides}')

A[0, 0] = 10
print('A:\n', A)
print('B:\n', B)  # Reflects the change
```

```python
x = np.array([10, 20, 30, 40, 50])
y = x[[0, 2, 3]]  # Copy: selected elements are gathered into a new array

x[0] = 999
print('x:', x)
print('y:', y)  # Unchanged
```
