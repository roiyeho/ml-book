```python
import numpy as np
```

Arrays with Fixed Values

```python
np.zeros((2, 3))
```

```python
np.ones((3, 2), dtype=int)
```

```python
np.full((2, 2), fill_value=7.5)
```

Arrays with Uninitialized Values

```python
np.empty((2, 4))
```

Arrays with Shape Based on Another Array

```python
x = np.array([[1, 2, 3], [4, 5, 6]])

np.zeros_like(x)
```

Arrays with Regular Spacing

```python
np.arange(0, 10, 2)
```

```python
np.arange(9).reshape(3, 3)
```

```python
np.linspace(0, 1, 5)
```

```python
np.logspace(-1, 2, 4)
```

```python
x = np.linspace(0, 1, 5)
y = np.linspace(1, 3, 3)
X, Y = np.meshgrid(x, y)

print(X)
print(Y)
```
