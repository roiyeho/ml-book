```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a grid of (x, y) points over the function domain
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Define a quadratic function: z = x^2 + 2xy + 3y^2
Z = X**2 + 2*X*Y + 3*Y**2

# Plot contour lines
contours = plt.contour(X, Y, Z, levels=15, colors='black', linewidths=0.5)

# Add labels to the contour lines
plt.clabel(contours, fontsize=8)

plt.title("Contour Plot of a Quadratic Function")
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(alpha=0.5)
plt.savefig('figures/contour_plot.pdf')
```
