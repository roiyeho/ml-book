```python
import plotly.express as px
import seaborn as sns

# Load Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Create interactive scatter plot
fig = px.scatter(
    iris,
    x='sepal_width',
    y='sepal_length',
    color='species',
    title="Iris Sepal Dimensions"
)
fig.show()
```
