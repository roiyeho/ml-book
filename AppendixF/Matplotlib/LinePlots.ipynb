```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)   
plt.savefig('figures/line_plot.pdf')
```

Format String

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = x**2

plt.plot(x, y, 'ro--')  # red ('r'), circle markers ('o'), dashed line ('--')
plt.savefig('figures/line_plot_fmt.pdf')
```

Multiple Lines

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.savefig('figures/multiple_lines.pdf')
```

Changing Figure Size

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.savefig('figures/figure_size.pdf')
```

Line Colors

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), color='magenta')
plt.plot(x, np.cos(x), color=(0.1, 0.6, 0.5))
plt.savefig('figures/line_colors.pdf')
```

Line Styles

```python
x = np.linspace(0, 10, 100)
styles = ['solid', 'dashed', 'dashdot', 'dotted']

plt.figure(figsize=(8, 4))

for i, style in enumerate(styles):
    y = np.sin(x + i)  # phase shift
    plt.plot(x, y, linestyle=style)

plt.savefig('figures/line_styles.pdf')
```

Line Width

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), linewidth=1)
plt.plot(x, np.sin(x + 0.5), linewidth=2)
plt.plot(x, np.sin(x + 1), linewidth=3)

plt.savefig('figures/line_width.pdf')
```

Markers

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)

plt.plot(x, y, 'o--')                 # Circle markers, dashed line
plt.plot(x, y + 0.5, marker='s')      # Square markers, solid line

plt.savefig('figures/markers.pdf')
```

Title and Axis Labels

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.title('Sine and Cosine', fontfamily='Comic Sans MS', fontsize=16, color='teal')
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y$', fontsize=12)

plt.savefig('figures/title_and_labels.pdf')
```

Legend

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), label='Sine')
plt.plot(x, np.cos(x), label='Cosine')
plt.legend()

plt.savefig('figures/legend.pdf')
```

Axes Limits

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlim(1, 2 * np.pi)
plt.ylim(-0.5, 1)
# Alternatively:
# plt.axis([1, 2 * np.pi, -0.5, 1])

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.savefig('figures/axes_limits.pdf')
```

Logarithmic Scales

```python
x = np.linspace(1, 10, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('log')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title("Semi-log plot")

plt.savefig('figures/semi_log_plot.pdf')
```

Equal Aspect Ratio

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.axis('equal')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.savefig('figures/equal_axis.pdf')
```

Custom Ticks

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)

ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(ticks, labels)
plt.yticks([-0.5, 0, 0.5, 1.0])

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.savefig('figures/custom_ticks.pdf')
```

Grid Lines

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.grid(True, alpha=0.5)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.savefig('figures/grid_lines.pdf')
```

Concluding Example

```python
x = np.linspace(0, 4 * np.pi, 400)
y1 = np.exp(-0.1 * x)
y2 = y1 * np.sin(x)

plt.plot(x, y1, label='$\exp(-0.1 x)$', color='tab:blue')
plt.plot(x, y2, label='$\exp(-0.1 x)\sin(x)$', color='tab:purple', linestyle='--', linewidth=2)

plt.title('Exponential Decay and Damped Sine Wave')
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi],
           ['0', '$\\pi$', '$2\\pi$', '$3\\pi$', '$4\\pi$'])
plt.grid(True, alpha=0.5)
plt.legend()

plt.savefig('figures/combined_line_plot.pdf')
```

```python
x = np.linspace(0, 4 * np.pi, 400)
y1 = np.exp(-0.1 * x)
y2 = y1 * np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label='$\exp(-0.1 x)$', color='tab:blue')
ax.plot(x, y2, label='$\exp(-0.1 x)\sin(x)$', color='tab:purple', linestyle='--', linewidth=2)

ax.set_title('Exponential Decay and Damped Sine Wave')
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$f(x)$', fontsize=12)
ax.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi])
ax.set_xticklabels(['0', '$\\pi$', '$2\\pi$', '$3\\pi$', '$4\\pi$'])
ax.grid(True, alpha=0.5)
ax.legend()
```
