```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an image from file
img = plt.imread('book_cover.jpg')

# Display the image
plt.imshow(img)
plt.axis('off')  # Hide the axes
plt.title("Book Cover")
plt.savefig('figures/rendering_images.pdf')
```

```python
# Extract RGB channels
red   = img[:, :, 0]
green = img[:, :, 1]
blue  = img[:, :, 2]

# Convert to grayscale (using a luminance-based weighted sum)
grayscale = 0.2989 * red + 0.587 * green + 0.114 * blue

# Flip horizontally
flipped = np.fliplr(grayscale)

# Display the result
plt.imshow(flipped, cmap='gray')
plt.title('Flipped Grayscale Image')
plt.axis('off')
plt.savefig('figures/image_processing.pdf')
```
