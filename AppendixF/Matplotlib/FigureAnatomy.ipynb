```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Create a figure and a single Axes just for drawing the schematic
fig, ax = plt.subplots(figsize=(8, 4))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_axis_off()

# =========================
# Figure rectangle
# =========================
fig_rect = Rectangle(
    (0.05, 0.10), 0.90, 0.80,
    fill=False, edgecolor='black', linewidth=1.8
)
ax.add_patch(fig_rect)
ax.text(0.5, 0.92, "Figure", ha='center', va='bottom', fontsize=12, fontweight='bold')

# =========================
# Axes rectangles (two subplots)
# =========================
# Left Axes
axes_left = Rectangle(
    (0.09, 0.16), 0.37, 0.65,
    fill=False, edgecolor='black', linewidth=1.4
)
ax.add_patch(axes_left)
ax.text(0.27, 0.825, "Axes", ha='center', va='bottom', fontsize=11, fontweight='bold')

# Right Axes
axes_right = Rectangle(
    (0.55, 0.16), 0.35, 0.65,
    fill=False, edgecolor='black', linewidth=1.4
)
ax.add_patch(axes_right)
ax.text(0.725, 0.825, "Axes", ha='center', va='bottom', fontsize=11, fontweight='bold')

# =========================
# Inside left Axes: data area + artists
# =========================
# Data area inside left Axes (inset rectangle)
data_area = Rectangle(
    (0.11, 0.19), 0.33, 0.53,
    fill=False, edgecolor='gray', linestyle='--', linewidth=1.0
)
ax.add_patch(data_area)

# "Axis" lines (bottom and left) inside the data area
# Bottom axis line
ax.plot([0.12, 0.43], [0.24, 0.24], color='black', linewidth=1.0)
# Left axis line
ax.plot([0.14, 0.14], [0.24, 0.72], color='black', linewidth=1.0)

# A sample "Line2D" inside the data area
x_line = [0.14, 0.22, 0.30, 0.38, 0.42]
y_line = [0.26, 0.50, 0.60, 0.68, 0.70]
ax.plot(x_line, y_line, color='tab:blue', linewidth=1.8)
ax.text(0.30, 0.55, "Line2D", fontsize=9, ha='left', va='bottom', color='tab:blue')

# A sample marker point
ax.scatter([0.22], [0.40], color='tab:red', s=25)
ax.text(0.225, 0.41, "Artist", fontsize=8, ha='left', va='bottom', color='tab:red')

# Axis labels as Text artists
ax.text(0.285, 0.23, "$x$-axis (Axis)", ha='center', va='top', fontsize=8)
ax.text(0.14, 0.48, "$y$-axis (Axis)", ha='right', va='center', rotation=90, fontsize=8)

# Title (Text)
ax.text(0.275, 0.73, "Title (Text)", ha='center', va='bottom', fontsize=9)

# Legend box inside data area
legend_box = Rectangle(
    (0.16, 0.60), 0.11, 0.10,
    fill=False, edgecolor='black', linewidth=1.0
)
ax.add_patch(legend_box)
ax.text(0.18, 0.68, "Legend", ha='left', va='top', fontsize=8)

# =========================
# Right Axes
# =========================
# Data area inside right Axes
data_area_r = Rectangle(
    (0.57, 0.19), 0.31, 0.53,
    fill=False, edgecolor='gray', linestyle='--', linewidth=1.0
)
ax.add_patch(data_area_r)

# Simple line in right axes
x_line_r = [0.58, 0.64, 0.70, 0.76, 0.86]
y_line_r = [0.28, 0.30, 0.50, 0.55, 0.60]
ax.plot(x_line_r, y_line_r, color='tab:orange', linewidth=1.5)

ax.text(0.725, 0.75, "Another Axes with its own data",
        ha='center', va='center', fontsize=8)

# Tight layout not really needed since axes are manual, but safe
plt.savefig('figures/figure_anatomy.pdf', bbox_inches='tight')
```
