```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
courses = ['ML', 'AI', 'CV', 'NLP', 'RL']
students = [120, 95, 75, 80, 60]

plt.bar(courses, students, color='skyblue', edgecolor='black')
plt.title("Enrollment by Course")
plt.xlabel("Course")
plt.ylabel("Number of Students")
plt.savefig('figures/bar_chart.pdf')
```

Horizontal Bar Chart

```python
plt.barh(courses, students, color='pink', edgecolor='black')
plt.title("Enrollment by Course")
plt.xlabel("Number of Students")
plt.ylabel("Course")
plt.savefig('figures/horizontal_bar_chart.pdf')
```

Stacked Bar Charts

```python
courses = ['ML', 'AI', 'CV', 'NLP', 'RL']
fall = [40, 30, 20, 35, 25]
spring = [50, 45, 25, 30, 20]

x = np.arange(len(courses))

plt.bar(x, fall, label='Fall', color='skyblue', edgecolor='black')
plt.bar(x, spring, bottom=fall, label='Spring', color='lightgreen', edgecolor='black')

plt.xticks(x, courses)
plt.ylabel("Number of Students")
plt.title("Stacked Bar Chart of Enrollment by Course and Semester")
plt.legend()
plt.savefig('figures/stacked_bar_chart.pdf')
```

Grouped Bar Charts

```python
courses = ['ML', 'AI', 'CV', 'NLP', 'RL']
fall = [40, 30, 20, 35, 25]
spring = [50, 45, 25, 30, 20]

x = np.arange(len(courses))
width = 0.35  # width of each bar

plt.bar(x - width/2, fall, width=width, label='Fall', color='skyblue', edgecolor='black')
plt.bar(x + width/2, spring, width=width, label='Spring', color='lightgreen', edgecolor='black')

plt.xticks(x, courses)
plt.ylabel("Number of Students")
plt.title("Grouped Bar Chart of Enrollment by Course and Semester")
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.savefig('figures/grouped_bar_chart.pdf')
```
