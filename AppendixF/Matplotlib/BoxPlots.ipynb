```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
group1 = np.random.normal(loc=0, scale=1, size=100)
group2 = np.random.normal(loc=1, scale=1.5, size=100)
group3 = np.random.normal(loc=-1, scale=0.8, size=100)

data = [group1, group2, group3]
plt.boxplot(data, tick_labels=['Group 1', 'Group 2', 'Group 3'], showmeans=True)

plt.title("Box Plot of Three Groups")
plt.ylabel("Value")
plt.grid(axis='y', alpha=0.5)
plt.savefig('figures/box_plot.pdf')
```
