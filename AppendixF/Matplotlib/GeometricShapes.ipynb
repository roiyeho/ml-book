```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')  # keep circles circular

rect = patches.Rectangle(
    (1, 1), width=3, height=2, edgecolor='blue', facecolor='skyblue'
)
ax.add_patch(rect)

circle = patches.Circle(
    (7, 3), radius=1.5, edgecolor='green', facecolor='lightgreen'
)
ax.add_patch(circle)

ellipse = patches.Ellipse(
    (7, 7), width=3, height=1.5, edgecolor='purple', facecolor='violet'
)
ax.add_patch(ellipse)

arrow = patches.FancyArrow(2, 7, dx=3, dy=-2, width=0.2, color='red')
ax.add_patch(arrow)

plt.title("Geometric Shapes")
plt.grid(alpha=0.5)
plt.savefig('figures/geometric_shapes.pdf')
```
