```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 50

# Random data
x = np.random.rand(n_points)
y = np.random.rand(n_points)
sizes = 100 * np.random.rand(n_points)  # Random sizes (scaled)
colors = np.random.rand(n_points)  # Random colors

plt.scatter(x, y, s=sizes, c=colors, cmap='Blues')
plt.colorbar()  # Show color scale
plt.title("Random Scatter Plot")
plt.savefig('figures/scatter_plot.pdf')
```
