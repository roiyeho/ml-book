```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 50

# Random 2D points (sampled uniformly from [0, 1))
x = np.random.rand(n_points)
y = np.random.rand(n_points)
sizes = 100 * np.random.rand(n_points)  # Random marker sizes (scaled)
colors = np.random.rand(n_points)       # Random scalar values for color mapping

plt.scatter(x, y, s=sizes, c=colors, cmap='Blues')
plt.colorbar()  # Colorbar for the scalar-to-color mapping
plt.title("Random Scatter Plot")
plt.savefig('figures/scatter_plot.pdf')
```
