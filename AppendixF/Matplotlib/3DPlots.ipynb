```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create 3D grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute the height values Z for the surface plot: z = sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='Blues')

# Labels
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$', labelpad=2)  # Prevents z-axis label from being cut off 
plt.title("3D Surface Plot")
plt.tight_layout()
plt.savefig('figures/3d_surface_plot.pdf', bbox_inches='tight')
```
