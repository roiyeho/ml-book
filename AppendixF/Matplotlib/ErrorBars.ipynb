```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)
error = np.abs(np.random.normal(loc=0.1, scale=0.05, size=len(x)))

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5, ecolor='red')
plt.title("Sine Wave with Error Bars")
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(alpha=0.5)

plt.savefig('figures/errorbar_plot.pdf')
```
