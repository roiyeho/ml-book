```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -15 * x**4 + 15 * x**2 + 3 * x

x = np.linspace(-1.0, 1.2, 100)
y = f(x) 
plt.plot(x, y)

# Approximate locations of the local optima
x_local_max = -0.65
x_local_min = -0.10

# Annotate the local maximum (left peak) 
plt.annotate("local maximum",
             xy=(x_local_max, f(x_local_max)),
             xytext=(x_local_max + 0.05, f(x_local_max) + 1.2),
             arrowprops=dict(arrowstyle='->'),
             ha='right', va='bottom')

# Annotate the local minimum (middle dip)
plt.annotate("local minimum",
             xy=(x_local_min, f(x_local_min)),
             xytext=(x_local_min + 0.15, f(x_local_min) - 1.4),
             arrowprops=dict(arrowstyle='->', color='red'),
             color='red', ha='left', va='top')

plt.xlabel('$x$')
plt.ylabel("$y$")
plt.tight_layout()
plt.savefig('figures/annotations.pdf')
```
