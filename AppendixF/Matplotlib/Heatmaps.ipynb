```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data with correlated features
np.random.seed(42)
size = 100
x1 = np.random.randn(size)
x2 = 0.5 * x1 + np.random.randn(size) * 0.3
x3 = np.random.randn(size)
x4 = x2 - 0.3 * x3 

# Stack variables into a matrix (shape: 4 x 100)
X = np.vstack([x1, x2, x3, x4])

# Compute the correlation matrix (shape: 4 x 4)
corr = np.corrcoef(X)

# Plot the heatmap
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')
plt.xticks(range(4), ['$x_1$', '$x_2$', '$x_3$', '$x_4$'])
plt.yticks(range(4), ['$x_1$', '$x_2$', '$x_3$', '$x_4$'])
plt.title('Correlation Matrix Heatmap')
plt.savefig('figures/heatmap.pdf')
```
