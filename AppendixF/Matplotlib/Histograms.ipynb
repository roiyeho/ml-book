```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Standard Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.savefig('figures/histogram.pdf')
```

2D Histogram

```python
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Correlation coefficient = 0.8
x, y = np.random.multivariate_normal(mean, cov, size=1000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
plt.colorbar()
plt.title("2D Histogram of Correlated Normal Variables")
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.savefig('figures/hist2d.pdf')
```
