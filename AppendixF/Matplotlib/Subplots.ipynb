```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sine")
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine")
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.savefig('figures/simple_subplots.pdf', bbox_inches='tight')
```

```python
x = np.linspace(0, 2 * np.pi, 100)
fig, axes = plt.subplots(2, 5, figsize=(10, 4), sharex=True, sharey=True)

for i, ax in enumerate(axes.flat):
    phase = i * np.pi / 5
    ax.plot(x, np.sin(x + phase))
    ax.set_title(f"Phase = ${i}\pi / 5$")

fig.suptitle("Sine Waves with Increasing Phase Shifts", fontsize=14)
plt.tight_layout()  # Adjust spacing to prevent overlapping labels
plt.savefig('figures/subplots_array.pdf', bbox_inches='tight')
```
