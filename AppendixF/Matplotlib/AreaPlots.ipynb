```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
delta = 0.2

plt.plot(x, y, label='Sine wave', color='blue')
plt.fill_between(x, y - delta, y + delta, color='blue', alpha=0.2, label='0.2 band')

plt.title("Sine Wave with Shaded Error Band")
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/fill_between.pdf')
```
