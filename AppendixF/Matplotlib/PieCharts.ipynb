```python
import matplotlib.pyplot as plt
```

```python
courses = ['ML', 'AI', 'CV', 'NLP', 'RL']
students = [120, 95, 75, 80, 60]

plt.pie(students, labels=courses, autopct='%1.1f%%', startangle=90)
plt.title("Course Enrollment Distribution")
plt.savefig('figures/pie_chart.pdf', bbox_inches='tight')
```
