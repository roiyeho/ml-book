```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x / 3)

fig, ax1 = plt.subplots()

# Plot on primary y-axis (left)
ax1.plot(x, y1, color='blue')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$\sin(x)$', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create twin y-axis (on the right) that shares the same x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', linestyle='--')
ax2.set_ylabel('$\exp(x/3)$', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Twin Y-Axes Example")
plt.savefig('figures/twin_axes.pdf')
```

Secondary Axes

```python
# Generate synthetic temperature data
x = np.linspace(0, 10, 100)  # e.g., time in hours
temp_celsius = 10 + 5 * np.sin(x)  # fluctuating temperatures

# Conversion functions
def celsius_to_fahrenheit(c):
    return c * 9 / 5 + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9

# Plotting
fig, ax = plt.subplots()
ax.plot(x, temp_celsius)
ax.set_xlabel("Time (hours)")
ax.set_ylabel("Temperature (C)")
ax.tick_params(axis='y')

# Add secondary y-axis on the right
secax = ax.secondary_yaxis('right',
            functions=(celsius_to_fahrenheit, fahrenheit_to_celsius))
secax.set_ylabel("Temperature (F)", color='tab:red')
secax.tick_params(axis='y', labelcolor='tab:red')

plt.title("Temperature Over Time")
plt.grid(alpha=0.5)
```
