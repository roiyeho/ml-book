```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
U = 2 * X
V = 2 * Y

plt.quiver(X, Y, U, V)
plt.title('Gradient Field of $f(x, y) = x^2 + y^2$')
plt.grid(alpha=0.5)

plt.savefig('figures/quiver_plot.pdf')
```
